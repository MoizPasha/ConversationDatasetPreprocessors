```python
import pandas as pd # import pandas for loading dataset
```

```python
# merge test , train and validation sets
train_df = pd.read_csv('/kaggle/input/dialogsum/CSV/train.csv')
test_df = pd.read_csv('/kaggle/input/dialogsum/CSV/test.csv')
val_df = pd.read_csv('/kaggle/input/dialogsum/CSV/validation.csv')
```

```python
# merge into one dataframe
df = pd.concat([train_df,test_df,val_df],ignore_index=True)
```

```python
# remove unnecssary data
df = df.drop(columns=['id','summary','topic'])
```

```python
x = df['dialogue'][0]
x
```

```python
import re

def extract_conversation(text):
    matches = re.findall(r"#(.*?)#:\s*(.*)", text)
    mID = 1
    messages = []
    for name, dialogue in matches:
        messages.append({
            'from': name,
            'message_id': mID,
            'message': dialogue
        })
        mID+=1
    return messages
        
```

```python
import json

output_data = {}
conv_id = 50000

for _,row in df.iterrows():
    conversation = extract_conversation(row['dialogue'])
    if (len(conversation) >= 8):
        output_data[conv_id] = conversation
        conv_id += 1
```
