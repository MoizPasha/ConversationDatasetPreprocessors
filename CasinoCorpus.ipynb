```python
from convokit import Corpus, download
corpus = Corpus(filename=download("casino-corpus"))
```

```python
import json
import emoji
from convokit import Corpus, download

# Initialize dictionary to hold all conversations
output_data = {}

# Custom Start number
conv_key = 10000

# Loop through conversations in corpus
for conversation in corpus.iter_conversations():
    messages = []
    
    for idx, utt in enumerate(conversation.iter_utterances(), start=1):
        messages.append({
            "from": utt.speaker.id,
            "message_id": idx,
            "message": emoji.replace_emoji(utt.text, replace='')
        })

    # Only keep conversations with at least 8 messages
    if len(messages) >= 8:
        output_data[str(conv_key)] = messages
        conv_key += 1
```
