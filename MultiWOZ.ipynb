```python
from datasets import load_dataset

dataset = load_dataset("multi_woz_v22")
```

```python
import json
from datasets import load_dataset

speaker_map = {0: 'USER', 1: 'SYSTEM'} # since multiwoz on hugging face has 0,1 

output_data = {}
conv_key = 20000  # Custom Start number

for split_name, split_data in dataset.items():
    for dialogue in split_data:
        if len(dialogue['turns']['turn_id']) >= 8:
            messages = []
            mID = 1
            for speaker, utterance in zip(dialogue['turns']['speaker'], dialogue['turns']['utterance']):
                messages.append({
                    "from": speaker_map[speaker],
                    "message_id": mID,
                    "message": utterance
                })
                mID += 1
            output_data[str(conv_key)] = messages
            conv_key += 1
```
